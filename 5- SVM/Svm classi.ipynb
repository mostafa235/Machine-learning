```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('J:/Data science/ML/Pyhon for ML/3.1 UNZIP_ME_FOR_NOTEBOOKS_V4/DATA/mouse_viral_study.csv')
```

```python
df.head()
```

```python
sns.scatterplot(data=df , x='Med_1_mL' , y='Med_2_mL',hue='Virus Present')
```

```python
x = df.drop('Virus Present',axis=1)
y = df['Virus Present']
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear' , C = 1000)
svm_model.fit(x,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(svm_model,x,y)
```

```python
model = SVC(kernel='linear' , C = 0.05)
model.fit(x,y)
```

```python
plot_svm_boundary(model,x,y)
```

```python
model = SVC(kernel='poly' , degree= 1, C=1)
model.fit(x,y)
plot_svm_boundary(model,x,y)
```

```python
model = SVC(kernel='poly' , degree= 2, C=1)
model.fit(x,y)
plot_svm_boundary(model,x,y)
```

```python
model = SVC(kernel='rbf' , C = 1)
model.fit(x,y)
plot_svm_boundary(model,x,y)
```

```python
model = SVC(kernel='rbf' , C=1 , gamma= 0.01)
model.fit(x,y)
plot_svm_boundary(model,x,y)
```

```python
model = SVC(kernel='rbf' , C=1 , gamma= 8)
model.fit(x,y)
plot_svm_boundary(model,x,y)
```

```python
model = SVC(kernel='rbf' , C = 1)
model.fit(x,y)
plot_svm_boundary(model,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = SVC()
param_grid = {'kernel':['linear','rbf'],
               'C':[0.01, 0.1, 1]}
grid_model = GridSearchCV(model,param_grid)
```

```python
grid_model.fit(x,y)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```
